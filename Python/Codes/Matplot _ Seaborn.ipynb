# Scatter, line & bar plots

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('Sales.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
avgsales = df.groupby('ProductCategory')['sales'].mean()
```

```python
avgsales
```

```python
x = list(avgsales.index)
y = list(avgsales.values)
```

```python
plt.title("Average Sales across Product Categories")
plt.plot(x, y, 'r') # r - red, k - black, m - magenta, c - cyan, w - white, y - yellow 
plt.xlabel("Product Categories")
plt.ylabel('Average Sales')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.title("Average Sales across Product Categories")
plt.plot(x, y, 'g') # r - red, k - black, m - magenta, c - cyan, w - white, y - yellow 
plt.xlabel("Product Categories")
plt.ylabel('Average Sales')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title("Average Sales across Product Categories")
plt.bar(x, y) # r - red, k - black, m - magenta, c - cyan, w - white, y - yellow 
# plt.xlabel("Product Categories")
# plt.ylabel('Average Sales')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title("Average Sales across Product Categories")
plt.bar(x, y, color = ['r', 'g', 'b', 'k', 'm']) # r - red, k - black, m - magenta, c - cyan, w - white, y - yellow 
# plt.xlabel("Product Categories")
# plt.ylabel('Average Sales')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize = (10, 5))
plt.title("Price Vs Sales")
plt.scatter(df['Price'][:100], df['sales'][:100])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title("Price Vs Sales")
plt.scatter(df['Price'][:100], df['sales'][:100], color = 'red', s = df['Price'][:100])
plt.show()
```

# Regplot, countplot, distplot

```python
df.columns
```

```python
plt.figure(figsize = (10, 5))
plt.title('Relationship between Price and Sales')
sns.regplot(x = "Price", y = "sales", data = df, color = "red")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.title('Frequency of Product Categories')
sns.countplot(x = "ProductCategory", data = df)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
# plt.title('Distribution of Sales')
sns.displot(x = "sales",  data = df, color = "blue", bins = 5, kde = True)
plt.show()
```
