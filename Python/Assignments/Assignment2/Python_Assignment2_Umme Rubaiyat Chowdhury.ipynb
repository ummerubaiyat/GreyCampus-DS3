```python
#importing the pandas library
import pandas as pd
#numpy used for mathematical functions
import numpy as np
#importing absenteesm dataset(provided)
df = pd.read_csv("Absenteeism_data+(2).csv")
```

3. Drop the first column “ID”, and save it back to “df”.

```python
#It drops the column name "ID", axis=1 means its a column, inpace= True means it will permanently
#deleted from the dataset
df.drop("ID",axis=1, inplace=True) 
df.head(2)
```

4. Drop the age column. And save it back to “df”

```python
#It will remove the  Age column from df
df.drop("Age",axis=1, inplace=True)
df.head(2)
```

5. What are the smallest and largest values of the column “reason for absence”

```python
#smallest values
print("Smallest Value of the column 'reason for absence': ",df["Reason for Absence"].min())
#largest values
print("Largest Value of the column 'reason for absence':",df["Reason for Absence"].max())
```

8. Create a line graph to show the relationship between the column “reason of absence” and “absenteesm time in hours”.

```python
x = df["Reason for Absence"]
y = df["Absenteeism Time in Hours"]
plt.figure(figsize=(30,20))#size
plt.title("Relationship between the column 'Reason of Absence' and 'Absenteesm Hime in hours'")#title
plt.plot(x,y,'b')#r=blue color
plt.xlabel("Reason of Absence")
plt.ylabel("Absenteesm Hime in hours")
plt.show()
```

9. Create a suitable graph to derive useful insights from the columns distance in hours and absenteeism hours

```python
x = df["Reason for Absence"]
y = df["Absenteeism Time in Hours"]

plt.figure(figsize=(10,5))
plt.title("Insights from the columns 'Distance to Work' and 'Absenteeism Time in Hours'")#title
value=(df["Reason for Absence"]) & (df["Absenteeism Time in Hours"])
df['color']=  np.where( value==True , "red", "green")
sns.regplot(x,y,data=df,scatter_kws={'facecolors':df['color']},line_kws={"color": "black"})
 
```

The graph avove is a regplot, where two columns points are represented with two different colors. Also, the black regression line is the optimal line for the two columns. 
It is visualy clear that there is no relationship between the two given columns.
