```python
#import the required libraries(you will have to change the path to the file according to your system to run this)
import pandas as pd 
import numpy as np
df = pd.read_csv("imus.csv") 
#to read the data you can convert and merge the data using code or excel. This might take some time if loading in excel format
```

# For Data in Stock

```python
#Some necessary steps to directly jump into solving the problem statements
```

```python
#To know the count of data
df["status"].value_counts()
```

```python
#If you have to remove duplicates
df = df.drop_duplicates()
```

```python
#To know the count of data after removing duplicates
df["status"].value_counts()
```

```python
options = ['In Stock']
#dataInStock
dis = df[df['status'].isin(options)] 
```

```python
len(dis)
```

 ## Which are the top Manufacturers with price, quantity,

```python
#To answer this question you need to get max price and qty rows 
maxPriceRow = dis[dis['price'] == dis['price'].max()]
print("Top Manufactuer with Price is", maxPriceRow['manufacturer'].to_string())
maxQuantityRow =  dis[dis['qty'] == dis['qty'].max()]
print("Top Manufactuer with Quantity is", maxQuantityRow['manufacturer'].to_string())
```

## Which are the most profitable products?

```python
#To answer this question you need toget rows of top 5 max priced products
profitableProducts = dis.sort_values(by=['price'], ascending = False).head()
profitableProducts
```

## Which is the product where you have seen decrease in the price?

```python
#To answer this question you need to get the products with different price and same description, category, manufacturer, partnumber
#get  products with different price and same manufacturer, partnumber
#since the length is same so a unique row can be indicated by manufacturer and part_number

print(len(dis[dis.duplicated( subset=['category',
            'manufacturer','part_number', 'sku', 'description','sku_create_date'],
                             keep=False)]))

print(len(dis[dis.duplicated( subset=['manufacturer','part_number'],
                             keep=False)]))
```

```python
#Now get  products with different price and same manufactuer, partnumber
df_diff = dis[dis.duplicated( subset=['manufacturer','part_number'],
                             keep=False)]
df_diff
#here you can see that df has values with different price
```

```python
#Also, group by manufacturer and part_number and get unique values of price
df_diff.groupby(['manufacturer','part_number'])['price'].nunique().head()

#We understand that a unique product has 2 different price value
```

```python
#So, if price in index1 > index 2 add the difference to a new column
df_diff['price_diff'] = df_diff.groupby(['manufacturer','part_number'])['price'].diff()
df_diff
```

```python
#then, get a new df with diff_price < 0 for decreasing in price value
lowPricedf = df_diff.loc[(df_diff['price_diff'] < 0 )]
print("Products with decrease in price are as follows:")
print(lowPricedf['manufacturer']+ ' ' + lowPricedf['part_number'])
```


## Which is the product where you have seen increase in the price?



```python
increased_price = df_diff.loc[(df_diff['price_diff'] > 0 )]
print("Products with increase in price are as follows:")
print(increased_price['manufacturer']+ ' ' + increased_price['part_number'])
```
