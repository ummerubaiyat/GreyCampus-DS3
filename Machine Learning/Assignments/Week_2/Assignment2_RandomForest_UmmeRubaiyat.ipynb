```python
#importing the libraries
import numpy as np
import pandas as pd
```

```python
#reading the csv data file
df=pd.read_csv('Position_Salaries.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Checking the aggregate values of the dataset
df.describe()
```

```python
#Assigning X and Y values
X=pd.DataFrame(df['Level'])
Y=df['Salary']
```

Checking the type and data of X and y variable

```python
type(X)
```

```python
type(Y)
```

```python
#Splitting train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
#importing RandomForest algorithm
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 20,random_state=0)
```

```python
#Fitting our model to train
model.fit(X_train, Y_train)
```

checking the accuracy of our model

```python
print((model.score(X_train, Y_train))*100)
print((model.score(X_test, Y_test))*100)
```

```python
model.predict([[6.5]]).reshape(1, 1)
```

We can conclude that, a person with level 6.5 will have a predicted salary of 1,50,000 which seems to be less that what the person has claimed to be 1,60,000. From here, we can conclude that the person is lying.
